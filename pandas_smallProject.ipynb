23. How many people have the word 'officer' in their job title? (This is pretty tricky)

```python
pay[pay['Job Titles'].str.contains('officer', case=False)]
```

Why use case=False?
By default, str.contains() is case-sensitive, so:

'Officer' 

'officer' 

'OFFICER' 

'oFfIcEr' 
only if you set case=False.

If you don't use case=False, then it will only match exactly as typed,


So, using case=False ensures you match all versions



###

Why use .shape[0]?
.shape[0] gives you the number of rows.

You're filtering the DataFrame with job titles containing 'officer', and then getting how many such rows exist.



```python
pay[pay['Job Titles'].str.contains('officer', case=False)].shape[0]
```
